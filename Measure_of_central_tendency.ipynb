# Mean

```python
#mean
#population = Î¼ = âˆ‘xi/N
#sample = xÌ„ =  âˆ‘xi/n
mean1 = df['math_score'].mean()
print(f"mean of math_score: {mean1}")

mean2 = df['reading_score'].mean()
print(f"mean of the reading_score: {mean2}")

mean3 = df['writing_score'].mean()
print(f"mean of the writing_score: {mean3}")
```

# Median

```python
#Median
#when no. is even then add middle two no.& divide by 2
#when no. is odd then take middle no.

median1 = df['math_score'].median()
print(f"mean of math_score: {median1}")

median2 = df['reading_score'].median()
print(f"mean of the reading_score: {median2}")

median3 = df['writing_score'].median()
print(f"mean of the writing_score: {median3}")
```

# Mode


```python
# mode
# frequent element/no occurs in data

mode1 = df['math_score'].mode()
print(f"mean of math_score: {mode1}")

mode2 = df['reading_score'].mode()
print(f"mean of the reading_score: {mode2}")

mode3 = df['writing_score'].mode()
print(f"mean of the writing_score: {mode3}")
```
